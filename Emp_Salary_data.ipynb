```python
for x in emp:
    row=x.split(",")
    salary=int(row[5])
    print("\n Total Salary for " + row[1] + " is: " + str(salary))
    DA=salary*0.25
    print("DA is" + str(DA))
    TZ=salary*0.15
    print(" TZ is" + str(TZ))
    HRA=salary*0.16
    print("HRA is" + str(HRA))
    CCA=salary*0.02
    print("CCA is" + str(CCA))
    MA=salary*0.05
    print("MA is"+str(MA))
    LIC=salary*0.05
    print("LIC is" + str(LIC))
    NETSalary=DA+TZ+HRA+HRA+CCA+MA+LIC
    print("NET-Salary for " + row[1] + " is: " + str(NETSalary))
print('\n')
```
