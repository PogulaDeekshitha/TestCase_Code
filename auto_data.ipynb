```python
# Represent Corelation between variables ?

sns.heatmap(x.corr(),annot=True)
```

```python
# plot different kinds of graphs for mean value represetation ? (use required headers) ?

sns.boxplot(x="origin",y="displacement",data=x)
```

```python
sns.violinplot(x="origin",y="acceleration",data=x)
```

```python
# Draw the pair plot using sns for mpg, weight,orign and with hue origin, set the size to 4 ?

sns.pairplot(
    x,
    x_vars=["mpg","weight"],
    y_vars=["mpg","weight"],hue="origin",size=4)
```
